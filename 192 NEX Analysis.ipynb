```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
```

```python
DISTS = list(range(1, 180))

ifort = {}
gfortran = {}
xlf = {}
xlf_strict = {}

for dist in DISTS:
    ifort[dist] = {}
    for comp in 'ZEN':
        ifort[dist][comp] = np.genfromtxt('test_ifort3/S%03d.SY.MX%s.sem.ascii' % (dist, comp))

    gfortran[dist] = {}
    for comp in 'ZEN':
        gfortran[dist][comp] = np.genfromtxt('test_gfortran3/S%03d.SY.MX%s.sem.ascii' % (dist, comp))

    xlf[dist] = {}
    for comp in 'ZEN':
        xlf[dist][comp] = np.genfromtxt('test_xlf3/S%03d.SY.MX%s.sem.ascii' % (dist, comp))

    xlf_strict[dist] = {}
    for comp in 'ZEN':
        xlf_strict[dist][comp] = np.genfromtxt('test_xlf_strict3/S%03d.SY.MX%s.sem.ascii' % (dist, comp))

    if dist % 10 == 0:
        print(dist)
```

```python
for dist in range(5, 180, 10):
    fig, axes = plt.subplots(3, 1, figsize=(16, 12), sharex=True)

    for ax, comp in zip(axes, 'ZEN'):
        ax.plot(ifort[dist][comp][:,0], ifort[dist][comp][:,1], label='ifort')
        ax.plot(gfortran[dist][comp][:,0], gfortran[dist][comp][:,1], label='gfortran')
        ax.plot(xlf[dist][comp][:,0], xlf[dist][comp][:,1], label='xlf -qnostrict')
        ax.plot(xlf_strict[dist][comp][:,0], xlf_strict[dist][comp][:,1], label='xlf -qstrict')
        ax.set_ylabel(comp)
        ax.legend(loc='best')

    if dist < 30:
        axes[2].set_xlim(ifort[dist]['Z'][0,0], 1000)
    elif 30 <= dist < 60:
        axes[2].set_xlim(2900, 3000)
    elif 60 <= dist < 90:
        axes[2].set_xlim(2000, 3000)
    elif 90 <= dist:
        axes[2].set_xlim(3000, ifort[dist]['Z'][-1,0])
    axes[2].set_xlabel('Time')
    fig.suptitle('%d deg, NEX=192' % (dist, ))
```

```python
dist = 45
for comp in 'ZEN':
    fig, axes = plt.subplots(3, 1, figsize=(16, 12), sharex=True)

    for ax, result, name in zip(axes, [gfortran, xlf, xlf_strict], ['gfortran', 'xlf -qnostrict', 'xlf -qstrict']):
        ax.plot(ifort[dist][comp][:,0], ifort[dist][comp][:,1], label='ifort')
        ax.plot(result[dist][comp][:,0], result[dist][comp][:,1], label=name)
        ax.set_ylabel(comp)
        ax.legend(loc='best')

    if dist < 30:
        axes[2].set_xlim(result[dist]['Z'][0,0], 1000)
    elif 30 <= dist < 60:
        axes[2].set_xlim(1000, 2000)
    elif 60 <= dist < 90:
        axes[2].set_xlim(2000, 3000)
    elif 90 <= dist:
        axes[2].set_xlim(3000, result[dist]['Z'][-1,0])
    axes[2].set_xlim(2900, 3000)
    axes[2].set_xlabel('Time')
    fig.suptitle('%d deg, NEX=192' % (dist, ))
```
